```python
import re

def extract_numbers(input_string):
    numbers = re.findall(r'\d+', input_string)
    return [int(num) for num in numbers]


def create_house_dict(filename):
    house_dict = {}
    small_amount = [] # список для тех, кого мало

    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:

          number = extract_numbers(line)[0]
          # Удаляем все цифры из строки
          clean_line = re.sub(r'\d+', '', line).strip()

          # Проверяем, что строка не пустая после удаления цифр
          if clean_line:
              # Разбиваем строку на части
              parts = clean_line.split()

              # Проверяем частоту этого дома
              if number > 7:
                  if len(parts) == 2:
                      house, house_name = parts
                      if house != 'House':
                          house_name = house
                      # Если дом уже есть в словаре, добавляем в список
                      if house_name in house_dict:
                          house_dict[house_name].append(clean_line)
                      else:
                          # Иначе создаем новый список
                          house_dict[house_name] = [clean_line]
              else:
                    small_amount.append(clean_line)


    if small_amount:
        house_dict['small_amount'] = small_amount



    return house_dict
```

```python
filename = 'houses.txt'
house_dictionary = create_house_dict(filename)
# Создаем обратный словарь для замены
reverse_house = {new_house: house for house, names in house_dictionary.items() for new_house in names}

# Заменяем значения в колонке name
data['house'] = data['house'].replace(reverse_house)
data_test['house'] = data_test['house'].replace(reverse_house)
```

```python
import re

def extract_numbers(input_string):
    numbers = re.findall(r'\d+', input_string)
    return [int(num) for num in numbers]


def create_title_dict(filename):
    title_dict = {}
    others = []
    small_amount = [] # список для тех, кого мало

    with open(filename, 'r', encoding='utf-8') as file:
          for line in file:
            number = extract_numbers(line)[0]
            # Удаляем все цифры из строки
            clean_line = re.sub(r'\d+', '', line).strip()

            # Проверяем, что строка не пустая после удаления цифр
            if clean_line:
                # Разбиваем строку на части
                parts = clean_line.split()
                title = parts[0]


                # Если фамилия уже есть в словаре, добавляем полное имя в список
                if title in title_dict:
                    title_dict[title].append(clean_line)
                elif number > 6:
                    # Иначе создаем новый список с полным именем
                    title_dict[title] = [clean_line]
                elif len(parts) > 1 and number < 7:
                    small_amount.append(clean_line)
                else:
                    others.append(clean_line)

    if others:
        title_dict['invalid_entries'] = others

    if small_amount:
        title_dict['small_amount'] = small_amount



    return title_dict
```

```python
filename = 'titles.txt'
title_dictionary = create_title_dict(filename)
# Создаем обратный словарь для замены
reverse_title = {new_title: title for title, names in title_dictionary.items() for new_title in names}

# Заменяем значения в колонке title
data['title'] = data['title'].replace(reverse_title)
data_test['title'] = data_test['title'].replace(reverse_title)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, shuffle=True, test_size=0.4, random_state=44)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = SGDRegressor()

param_grid = {
        'penalty': ['l2', 'l1', 'elasticnet', None],
        'loss': ['epsilon_insensitive', 'squared_epsilon_insensitive', 'huber', 'squared_error']
}

grid_search = GridSearchCV(model,
                           param_grid,
                           cv=7,
                           n_jobs=-1,
                           verbose=2)

# запустим поиск
grid_search.fit(X_train_scaled, y_train)

# выведем наилучшие параметры
print(grid_search.best_params_)

best_model_for_age = grid_search.best_estimator_

print(f'MSE train = {mean_squared_error(y_train, best_model_for_age.predict(X_train_scaled))}')
print(f'MSE test = {mean_squared_error(y_test, best_model_for_age.predict(X_test_scaled))}')
print(f'MAE train = {mean_absolute_error(y_train, best_model_for_age.predict(X_train_scaled))}')
print(f'MAE test = {mean_absolute_error(y_test, best_model_for_age.predict(X_test_scaled))}')
```

```python
# Расчет метрик
y_pred = best_model_for_age.predict(X_test_scaled)


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # Корень из MSE
r2 = r2_score(y_test, y_pred)

# Вывод результатов
print(f'Средняя абсолютная ошибка (MAE): {mae:.4f}')
print(f'Средняя квадратичная ошибка (MSE): {mse:.4f}')
print(f'Корень из средней квадратичной ошибки (RMSE): {rmse:.4f}')
print(f'Коэффициент детерминации (R²): {r2:.4f}')
```

`LogisticRegression`


```python
model = LogisticRegression()

param_grid = {
        'C': [0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs','liblinear', 'newton-cg',
                   'newton-cholesky', 'sag','saga'],
        'max_iter': [10, 50, 75, 100],
}

grid_search = GridSearchCV(model,
                           param_grid,
                           cv=7,
                           n_jobs=-1,
                           verbose=2)

# запустим поиск
grid_search.fit(X, y)

# выведем наилучшие параметры
print(grid_search.best_params_)

logistic_regression = grid_search.best_estimator_

print(f'MSE train = {mean_squared_error(y_train, logistic_regression.predict(X_train))}')
print(f'MSE test = {mean_squared_error(y_test, logistic_regression.predict(X_test))}')
print(f'MAE train = {mean_absolute_error(y_train, logistic_regression.predict(X_train))}')
print(f'MAE test = {mean_absolute_error(y_test, logistic_regression.predict(X_test))}')
```
